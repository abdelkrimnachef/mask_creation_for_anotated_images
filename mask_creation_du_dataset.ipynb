```python
img={}
for folders in  os.listdir(path2l) : 
  
  x=open(path2l+'/'+folders, "r")
  new_key= []
  
  name=folders.split('.')[0]
  new_key.append(name)

  img[new_key[0]]=(x.read())
print(img)
```

```python
img1={}
for i in img.keys():
  dict_values=[]
  for j in img[i].split('Vehicle registration plate '):
    
    dict_values.append(j)
    img1.update({i:dict_values})
   
print(img1)
```

```python
df = pd.DataFrame.from_dict(img1, orient ='index') 
df.reset_index(level=0, inplace=True)
df
```

```python
df = df.replace(r'\n',' ', regex=True) 
df
```

```python
next_df = pd.DataFrame(columns=['names','points'])
next_df['names']= df['index']
next_df['points']=list(df[1])
next_df
```

```python
for i in df:
  if i!=0 and i!='index' and i!=1:
    for j in df[i]:
      x=[]
      if j!= None:
        x=list(df[df[i]==j].index)
        x.append(j)
        x[0]=df.loc[x[0],'index']
        next_df.loc[len(next_df)+1,['names']]=x[0]
        next_df.loc[len(next_df),['points']]=x[1]
```

```python
list_images=[]
for index, row in df2.iterrows():
  img_name=df2.loc[index, "names"]   
  img_name_jpg=img_name+'.jpg'  
  img=cv2.imread(path2+'/'+img_name+'.jpg')
  
  img_points=df2.loc[index, "points"]
  points=img_points.split(" ")
  list_points=[]
  for i in points:
    if i !="":
      list_points.append(i)
  start_point =(int(float(list_points[0])),int(float(list_points[1])))
  end_point =(int(float(list_points[2])),int(float(list_points[3])))
  color=(255,255,255)
  
  
  if img_name_jpg not in list_images:
      fake_image = np.zeros(img.shape)
      fake_image = cv2.rectangle(fake_image, start_point, end_point, color, -1)
  
      cv2.imwrite(path2e+'/'+img_name+'.jpg',fake_image)
  elif img_name_jpg in list_images:
      fake_image =cv2.imread(path2e+'/'+img_name+'.jpg')
  
      fake_image = cv2.rectangle(fake_image, start_point, end_point, color, -1)
  
      cv2.imwrite(path2e+'/'+img_name+'.jpg',fake_image)

  list_images.append(img_name+'.jpg')
```
